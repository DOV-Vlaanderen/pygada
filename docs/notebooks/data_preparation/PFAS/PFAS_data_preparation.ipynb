# PFAS data preparation

```python
import pygada.data_preparation.data_cleaning.PFAS_data_cleaning as PFAS_data_cleaning
import pandas as pd
from pygada.data_preparation.data_filtering.data_wrangling import FilterTransformData
from pygada.data_preparation.data_filtering.unique_xyz import filter_to_3d
```

## Data cleaning

The output consist out of three different results:
- A dataframe with the 'error' records
- A dataframe with the 'clean' records
- A metadata file

If 'save' is not specified the results will be saved.

### Combined soil dataset

```python
# Get the combined soil results of data cleaning

combined_soil_results = PFAS_data_cleaning.soil('../../../../pygada/datasets/PFAS/gecombineerde_bodem_dataset.csv', save=False)
```

```python
# Get the error dataset

combined_soil_error = combined_soil_results[0]
combined_soil_error.info()
combined_soil_error
```

```python
# Get the clean dataset

combined_soil_clean = combined_soil_results[1]
combined_soil_clean.info()
combined_soil_clean
```

```python
# Get the soil metadata

combined_soil_metadata = combined_soil_results[2]
print(combined_soil_metadata)
```

### Combined groundwater dataset

```python
# Get the combined groundwater results of data cleaning

combined_groundwater_results = PFAS_data_cleaning.groundwater('../../../../pygada/datasets/PFAS/gecombineerde_grondwater_dataset.csv', save=False)
```

```python
# Get the error dataset

combined_groundwater_error = combined_groundwater_results[0]
combined_groundwater_error.info()
combined_groundwater_error
```

```python
# Get the clean dataset

combined_groundwater_clean = combined_groundwater_results[1]
combined_groundwater_clean.info()
combined_groundwater_clean
```

```python
# Get the groundwater metadata

combined_groundwater_metadata = combined_groundwater_results[2]
print(combined_groundwater_metadata)
```

### Create one dataset

```python
combined_soil_clean['matrix'] = 'Soil'
combined_groundwater_clean['matrix'] = 'Groundwater'

combined_dataset = pd.concat([combined_soil_clean, combined_groundwater_clean])
combined_dataset = combined_dataset.reset_index(drop=True)
combined_dataset
```

## Data filtering

### Data wrangling

```python
# Initialize the class

ftd = FilterTransformData(combined_dataset)
```

#### Filter on attribute

```python
combined_dataset_filtered_attribute = ftd.filter_on_attribute()
combined_dataset_filtered_attribute
```

#### Transform necessary columns to floats

```python
combined_dataset_filtered_float = ftd.transform_to_float()
combined_dataset_filtered_float
```

#### Check units

```python
combined_dataset_filtered_float_unique_unit = combined_dataset_filtered_float.copy()
combined_dataset_filtered_float_unique_unit = combined_dataset_filtered_float_unique_unit.unit.value_counts(dropna=False)
combined_dataset_filtered_float_unique_unit
```

```python
combined_dataset_filtered_unit = ftd.check_units()
combined_dataset_filtered_unit
```

```python
combined_dataset_filtered_unit_unique_unit = combined_dataset_filtered_unit.copy()
combined_dataset_filtered_unit_unique_unit = combined_dataset_filtered_unit_unique_unit.unit.value_counts(dropna=False)
combined_dataset_filtered_unit_unique_unit
```

### 3D filtering

```python
df = filter_to_3d(combined_dataset_filtered_unit)
df
```

```python
df.to_csv('../../../../pygada/datasets/PFAS/demo_dataset_filtered_3D.txt', encoding='utf-8-sig')
```
