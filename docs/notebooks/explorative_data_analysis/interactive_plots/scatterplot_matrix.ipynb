## Import Dependencies

```python
from highcharts_core.chart import Chart
import pandas as pd
import numpy as np
```

## Prepare the Configuration

## Populate the Dataframe

```python
df = pd.read_csv('C:/Users/vandekgu/OneDrive - Vlaamse overheid - Office 365/Documenten/PycharmProjects/pygada/pygada/test_data/results/PFAS_gw_VMM.csv')

df = df.pivot(index='index', columns='parameter', values='value')
df = df[['PFOS', 'PFHxS', 'PFNA', 'PFOA']]
df = np.log(df[:]).replace(np.NINF, np.nan)  # todo: check if necessary
df = df.dropna()
df
```

## Create the Series

```python
data_highcharts = []
columns = list(df.columns)

for i in range(len(columns)):
    for k in range(len(columns)):
        scatter_data = []
        for j in range(len(df)):
            scatter_data.append([df[columns[i]][j].round(2), df[columns[k]][j].round(2)])

        data_highcharts.append({"name":f'{columns[i]}-{columns[k]}', "data":scatter_data, "type": 'scatter', 'xAxis': i, 'yAxis': k})

data_highcharts
```

```python
count = int(len(columns))+1
print(count)

yAxis = [{
    'title': {
      'text': f'{columns[0]}'
    },
    'height': f'{int(100/count)}%',
  }]
xAxis = [{
    'title': {
      'text': f'{columns[0]}'
    },
    'width': f'{int(100/count)}%',
  }]

for i in range(1,count-1):
    yAxis.append({
    'title': {
      'text': f'{columns[i]}'
    },
    'top': f'{((i*int(100/count))+i*(int(100/count)/(count-2)))}%',
    'height': f'{int(100/count)}%',
    'offset': 0,
  })

    xAxis.append({
    'title': {
      'text': f'{columns[i]}'
    },
    'left': f'{(i*int(100/count)+i*int(100/count)/(count-2))}%',
    'width': f'{int(100/count)}%',
    'offset': 0,
  })

yAxis
```

```python

options_kwargs = {

    'title': {
        'text': 'Scatterplot of parameters'
    },

    'legend': {
        'enabled': False
    },

  'yAxis': yAxis,

  'xAxis': xAxis,

    "tooltip": {
        "pointFormat": '{point.series.name} <br/> {point.x}-{point.y}'
    },

}
```

## Assemble the Chart

```python
chart = Chart(options = options_kwargs)
chart.add_series(*data_highcharts)
```

## Display the Chart

```python
chart.display(container = 'highcharts_container')
```
