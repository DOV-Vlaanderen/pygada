```python
# Get the unique parameters
pd.set_option('display.max_rows', None)
df_soil_unique_parameters = df_soil_original.copy()
df_soil_unique_parameters_count = df_soil_unique_parameters.parameter.value_counts(dropna=False)
print(df_soil_unique_parameters_count)
```

The sum parameters are obtained based on the 'lower bound' principle. This implies that when a value is below the detection limit it is set as 0.
To not have this interpretation already made, these parameters are dropped from the dataset.

- totaal PFAS indicatief
- totaal PFAS
- totaal PFAS kwantitatief
- PFAS Som kwantitatief
- PFAS Som indicatief
- PFAS totaal (Kwantitatief en indicatief)
- EFSA-4

Some parameters are the same, but have different names.

- 'PFOAtotal' and 'som PFOA'
- 'PFOStotal' and 'som PFOS'
- 'DONA' and 'ADONA'

## Data wrangling

```python
df_soil_original_result = df_soil_original.copy()
df_soil_error = pd.DataFrame()
original_columns = df_soil_original.columns.values.tolist()
df_soil_error = pd.DataFrame(columns=original_columns)
df_error_count = []

# Drop rows with empty `date`
df_soil_result_1 = df_soil_original_result[(~df_soil_original_result['date'].isna())]
df_soil_error_date = df_soil_original_result[(df_soil_original_result['date'].isna())]
df_soil_error_date['error_date'] = 'No date'
df_soil_error = df_soil_error.merge(df_soil_error_date, how='outer', on=original_columns)
df_error_count.append(len(df_soil_error_date))

# Drop rows with sum parameters
df_soil_result_2 = df_soil_result_1[(~df_soil_result_1['parameter'].isin(['totaal PFAS indicatief', 'totaal PFAS', 'totaal PFAS kwantitatief', 'PFAS Som kwantitatief', 'PFAS Som indicatief', 'PFAS totaal (Kwantitatief en indicatief)', 'EFSA-4']))]
df_soil_error_sum_parameter = df_soil_original_result[(df_soil_original_result['parameter'].isin(['totaal PFAS indicatief', 'totaal PFAS', 'totaal PFAS kwantitatief', 'PFAS Som kwantitatief', 'PFAS Som indicatief', 'PFAS totaal (Kwantitatief en indicatief)', 'EFSA-4']))]
df_soil_error_sum_parameter['error_sum_parameter'] = 'Sum parameter'
df_soil_error = df_soil_error.merge(df_soil_error_sum_parameter, how='outer', on=original_columns)
df_error_count.append(len(df_soil_error_sum_parameter))

# Drop rows with higher than detection condition
df_soil_result_3 = df_soil_result_2[(~df_soil_result_2['detection_condition'].isin(['>']))]
df_soil_error_dc = df_soil_original_result[(df_soil_original_result['detection_condition'].isin(['>']))]
df_soil_error_dc['error_detection_condition'] = 'Higher than detection condition'
df_soil_error = df_soil_error.merge(df_soil_error_dc, how='outer', on=original_columns)
df_error_count.append(len(df_soil_error_dc))

# Drop rows with unrealistic units
df_soil_result_4 = df_soil_result_3[(~df_soil_result_3['unit'].isin(['ng/l', 'µg/l', 'µg Sn/kg ds', 'µg/kg']))]
df_soil_error_unit = df_soil_original_result[(df_soil_original_result['unit'].isin(['ng/l', 'µg/l', 'µg Sn/kg ds', 'µg/kg']))]
df_soil_error_unit['error_unrealistic_unit'] = "Unrealistic unit"
df_soil_error = df_soil_error.merge(df_soil_error_unit, how='outer', on=original_columns)
df_error_count.append(len(df_soil_error_unit))

# Replace parameter values
df_soil_result_4['parameter'] = df_soil_result_4['parameter'].replace('som PFOA', 'PFOAtotal')
df_soil_result_4['parameter'] = df_soil_result_4['parameter'].replace('som PFOS', 'PFOStotal')
df_soil_result_4['parameter'] = df_soil_result_4['parameter'].replace('ADONA', 'DONA')

df_soil_result_4.to_csv('../../data_input/PFAS_concentrations/combined_soil_dataset_cleaned.txt')
df_soil_error.to_csv('../../data_input/PFAS_concentrations/combined_soil_dataset_error.txt')
df_error_count = pd.DataFrame({'Error':['date', 'sum parameter', 'detection condition', 'unrealistic unit'], 'Count':df_error_count})
```

```python
pd.set_option('display.max_rows', None)
print(f'{df_soil_result_4.info()}\n\nClean dataset example\n{df_soil_result_4.head()}\n\nCount parameters\n{df_soil_result_4.parameter.value_counts(dropna=False)}\n\nCount detection conditions\n{df_soil_result_4.detection_condition.value_counts(dropna=False)}\n\nCount units\n{df_soil_result_4.unit.value_counts(dropna=False)}\n\nCount source\n{df_soil_result_4.source.value_counts(dropna=False)}')
```
